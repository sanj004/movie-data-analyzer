```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("tmdb_5000_movies.csv")

print("Shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isna().sum())
```

```python
df = df.drop(columns=['homepage','tagline'])

df = df.dropna(subset=['overview','release_date','runtime'])

df = df.drop_duplicates()

print(df.shape)
```

```python
long_and_rated = df[(df['runtime'] > 120) & (df['vote_average'] > 7.0)]
print("Shape:", long_and_rated.shape)
print("\n First five rows:")
display(long_and_rated.head())
```

```python
df['profit'] = df['revenue'] - df['budget']
df['roi'] = (df['profit'] /df['budget']) * 100
display(df[['title','budget','revenue','profit','roi']].head())
```

```python
print("Rating mean:",np.mean(df['vote_average']))
print("Rating median:",np.median(df['vote_average']))
print("Rating std deviation:", np.std(df['vote_average']))

print("Profit mean:", np.mean(df['profit']))
print("Profit median:", np.median(df['profit']))
print("Profit std deviation:", np.std(df['profit']))

clean_roi = df[df['budget'] > 0]['roi']

print("ROI mean:", np.mean(clean_roi))
print("ROI median:", np.median(clean_roi))
print("ROI std deviation:", np.std(clean_roi))
```

```python
top10_profit = df.sort_values(by='profit', ascending=False)[['title','profit','roi']].head(10)
top10_rated = df.sort_values(by='vote_average', ascending=False)[['title','vote_average','profit']].head(10)

display(top10_profit)
display(top10_rated)
```

```python
df['main_genre'] = df['genres'].apply(lambda x: eval(x)[0]['name'] if len(eval(x)) > 0 else 'Unknown')

genre_stats = df.groupby('main_genre')[['vote_average','roi']].mean()

genre_stats = genre_stats.sort_values(by='vote_average', ascending=False)

display(genre_stats)
```

```python
genre_stats['vote_average'].plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Avg Rating')
plt.title('Avg movie rating by Gnere')
plt.xticks(rotation=45)
plt.show()
```
